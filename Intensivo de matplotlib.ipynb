```python
import matplotlib.pyplot as plt
from cycler import cycler
# Dados fictícios
plt.style.use('bmh')
meses = range(1, 13)
eletronicos = [100, 120, 110, 130, 150, 140, 160, 170, 180, 200, 190, 210]
roupas = [80, 90, 100, 110, 120, 130, 140, 130, 120, 110, 100, 90]
alimentos = [150, 160, 170, 180, 190, 200, 210, 220, 230, 240, 250, 260]
             
print(min(eletronicos)) 
#Grafico de Linhas
plt.figure(figsize=(15,5))    
plt.plot(meses,eletronicos ,label='Vendas de Eletronicos ao longo do mes',marker='o')
plt.plot(meses,roupas ,label='Vendas de roupas ao longo do mes',marker='o')
plt.plot(meses,alimentos ,label='Vendas de alimentos ao longo do mes',marker='o')
plt.grid(True)
plt.axvspan(7,10,color='lightgreen')
plt.text(7,110,'lucros crescentes(Eletronicos,alimentos)',weight='bold',fontsize=8)
plt.xlabel('Mes')
plt.ylabel('Vendas em milhoes ($)')
plt.annotate('Menor Valor',color='red',xy=(1,83),arrowprops=dict(facecolor='red'),xytext=(2,110))
plt.legend()
plt.show()


#Grafico de barras
cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color',cores)
plt.rc('axes',prop_cycle=ciclo_cores)
plt.figure(figsize=(10, 5))
barras_eletronicos = plt.bar(meses, eletronicos, label='Eletrônicos')
barras_roupas = plt.bar(meses, roupas, label='Roupas', bottom=eletronicos)
barras_alimentos = plt.bar(meses, alimentos, label='Alimentos', bottom=[eletronicos[i] + roupas[i] for i in range(len(meses))])
plt.bar_label(barras_eletronicos,label_type='center',fontweight='bold')
plt.bar_label(barras_roupas,label_type='center',fontweight='bold')
plt.bar_label(barras_alimentos,label_type='center',fontweight='bold')
plt.title('Barras emplihadas das vendas ao longo dos meses',fontweight='bold',fontsize='15')
plt.xlabel('meses')
plt.ylabel('vendas')
plt.tick_params(axis='x',length=0)
plt.tick_params(axis='y',length=0)
plt.grid(False)
plt.xticks(ticks=meses)
plt.legend()
plt.show()


#Grafico de Pizza
plt.figure(figsize=(10, 5))
total_vendas_eletronicos = sum(eletronicos)
total_vendas_roupas = sum(roupas)
total_vendas_alimentos =  sum(alimentos)
total_vendas = [total_vendas_eletronicos, total_vendas_roupas,total_vendas_alimentos]
categorias = ['Eletrônicos', 'Roupas', 'Alimentos']
plt.pie(total_vendas, labels=categorias, autopct='%1.1f%%', startangle=140)
plt.title('porcentagem(%) em participação de vendas entre os ramos de vendas',fontweight='bold')
plt.show()

#Grafico de disperção
plt.figure(figsize=(8, 6))
plt.scatter(meses, eletronicos, label='Vendas de Eletrônicos', color='blue', marker='o')
plt.scatter(meses, roupas, label='Vendas de Roupas', color='red', marker='s')
plt.scatter(meses, alimentos, label='Vendas de Alimentos', color='green', marker='^')
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.title('Dispersão de Vendas por Categoria ao longo do Tempo')
plt.legend()
plt.grid(True)
plt.xticks(ticks=meses)
plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler

# Dados fictícios
plt.style.use('bmh')
meses = range(1, 13)
eletronicos = [100, 120, 110, 130, 150, 140, 160, 170, 180, 200, 190, 210]
roupas = [80, 90, 100, 110, 120, 130, 140, 130, 120, 110, 100, 90]
alimentos = [150, 160, 170, 180, 190, 200, 210, 220, 230, 240, 250, 260]

fig, eixos = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Gráfico de Linhas
eixos[0, 0].plot(meses, eletronicos, label='Vendas de Eletrônicos ao longo do mês', marker='o')
eixos[0, 0].plot(meses, roupas, label='Vendas de roupas ao longo do mês', marker='o')
eixos[0, 0].plot(meses, alimentos, label='Vendas de alimentos ao longo do mês', marker='o')
eixos[0, 0].grid(True)
eixos[0, 0].axvspan(7, 10, color='lightgreen')
eixos[0, 0].text(7, 110, 'lucros crescentes(Eletronicos,alimentos)', weight='bold', fontsize=8)
eixos[0, 0].set_xlabel('Mês')
eixos[0, 0].set_ylabel('Vendas em milhões ($)')
eixos[0, 0].annotate('Menor Valor', color='red', xy=(1, 83), arrowprops=dict(facecolor='red'), xytext=(2, 110))
eixos[0, 0].legend()

# Gráfico de Barras
cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)
barras_eletronicos = eixos[0, 1].bar(meses, eletronicos, label='Eletrônicos')
barras_roupas = eixos[0, 1].bar(meses, roupas, label='Roupas', bottom=eletronicos)
barras_alimentos = eixos[0, 1].bar(meses, alimentos, label='Alimentos',
                                   bottom=[eletronicos[i] + roupas[i] for i in range(len(meses))])
eixos[0, 1].bar_label(barras_eletronicos, label_type='center', fontweight='bold')
eixos[0, 1].bar_label(barras_roupas, label_type='center', fontweight='bold')
eixos[0, 1].bar_label(barras_alimentos, label_type='center', fontweight='bold')
eixos[0, 1].set_title('Barras empilhadas das vendas ao longo dos meses', fontweight='bold', fontsize='15')
eixos[0, 1].set_xlabel('Meses')
eixos[0, 1].set_ylabel('Vendas')
eixos[0, 1].tick_params(axis='x', length=0)
eixos[0, 1].tick_params(axis='y', length=0)
eixos[0, 1].grid(False)
eixos[0, 1].set_xticks(ticks=meses)
eixos[0, 1].legend()

# Gráfico de Pizza
total_vendas_eletronicos = sum(eletronicos)
total_vendas_roupas = sum(roupas)
total_vendas_alimentos = sum(alimentos)
total_vendas = [total_vendas_eletronicos, total_vendas_roupas, total_vendas_alimentos]
categorias = ['Eletrônicos', 'Roupas', 'Alimentos']
eixos[1, 0].pie(total_vendas, labels=categorias, autopct='%1.1f%%', startangle=140)
eixos[1, 0].set_title('Porcentagem (%) em participação de vendas entre os ramos de vendas', fontweight='bold')

# Gráfico de Dispersão
eixos[1, 1].scatter(meses, eletronicos, label='Vendas de Eletrônicos', color='blue', marker='o')
eixos[1, 1].scatter(meses, roupas, label='Vendas de Roupas', color='red', marker='s')
eixos[1, 1].scatter(meses, alimentos, label='Vendas de Alimentos', color='green', marker='^')
eixos[1, 1].set_xlabel('Meses')
eixos[1, 1].set_ylabel('Vendas')
eixos[1, 1].set_title('Dispersão de Vendas por Categoria ao longo do Tempo')
eixos[1, 1].legend()
eixos[1, 1].grid(True)
eixos[1, 1].set_xticks(ticks=meses)

plt.tight_layout()
plt.show()
```
